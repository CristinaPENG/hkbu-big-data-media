# 3. Price and Rating 
### Compared the price with rating, it can be found out that items with price range from 20 dollars or 30 dollars tend to receive higher stars. Items with higher price may be expected with high quality, thus leading to a lower rating. Items with its price in the highest level, however, would not receive rating lower than 3.5 stars.

```python
# clean the rating data into int rather than str.
list_rate = []
number = 0
for word in df['Rate']:
    rate = [df['Rate'][number].split()[0]]
    list_rate.append(rate)
    number = number + 1
```

```python
list_rate
```

```python
#price and rate
price = df['Price']
rate = list_rate
sValue = price * 10
plt.scatter(price, rate, marker = 's',c = np.random.rand(96), s= sValue, alpha = 0.5)
plt.title("Price And Rate")  
plt.xlabel("Price")  
plt.ylabel("Rating") 
plt.show()
```
