```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Amazon_dress.csv')
df
```

```python
#the most expensive dress($)
df[df['Price']==df['Price'].max()]  
```

```python
#brands' appearance times
fig = plt.figure(figsize = (96,7))
Brand_counts = df['Brand'].value_counts()
plt.bar(Brand_counts.index, Brand_counts)
```

```python
#price and rate
price = df['Price']
rate = df['Rate']
sValue = price * 10
plt.scatter(price, rate, marker = 's',c = np.random.rand(96), s= sValue, alpha = 0.5)
plt.title("Price And Rate")  
plt.xlabel("Price")  
plt.ylabel("Rate") 
plt.show()
```
