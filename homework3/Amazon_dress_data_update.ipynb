```python
import requests
import csv
from bs4 import BeautifulSoup
```

```python
def get_the_page(url):  
    r=requests.get(url)
    mypage=BeautifulSoup(r.text)
    
    list_name = []
    titles = mypage.find_all('li')
    for title in titles:
        name = title.find('h2')
        if name is None:
            continue
        list_name.append(name.text)
    
    list_price = []
    numbers = mypage.find_all('div',attrs = {'class':'a-row a-spacing-none'})
    for number in numbers:
        price = number.find('span',attrs = {'class':'sx-price-whole'})
        if price is None:
            continue
        list_price.append(price.text)
        
    list_brand = []
    brands = mypage.find_all('div',attrs = {'class':'a-column a-span12 a-text-center'})
    for brand in brands:
        brand_name = brand.find('span',attrs = {'class':'a-color-secondary s-overflow-ellipsis s-size-mild'})
        if brand_name is None:
            continue
        list_brand.append(brand_name.text)
        
    list_rate = []
    stars = mypage.find_all('a',attrs = {'class':'a-popover-trigger a-declarative'}) #href="javascript:void(0)"
    for star in stars:
        rate = star.find('span',attrs = {'class':'a-icon-alt'})
        if rate is None:
            continue
        list_rate.append(rate.text)
        
    page_content = []
    for i in range (0,len(list_name)):
        page_content.append([list_name[i],list_price[i],list_brand[i],list_rate[i]])
    
    return page_content
```

```python
url_list=[]
for i in range(0,50):
    page_number=str(i+1)
    url = 'https://www.amazon.com/s/ref=lp_10886357011_pg_'+ page_number +'?rh=n%3A7141123011%2Cn%3A10886357011%2Ck%3Adress&page='+ page_number +'&keywords=dress&ie=UTF8&qid=1522039975'
    url_list.append(url)
```

```python
information=[]
for item in url_list:
    my = get_the_page(item)
    information.extend(my)
```

```python
with open('Amazon_dress.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    header= ['Name','Price','Brand','Rate']
    spamwriter.writerow(header)
    spamwriter.writerows(information)
```
